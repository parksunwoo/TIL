#### [문제 6] 2016년에는 서울시 일부 지역에서 따릉이 사업을 시작하였고 2017년에 들어서서 서울시내 모든 구로 확대하여 시행하고 있다.  구별 연간 자전거 이용 거리의 총 합계가 연도별로 차이가 있는지 확인하기 위하여 2016년 및 2017년도 2년 모두 따릉이 사업을 시행한 12개 구에 대하여 대응 표본 t검정을 시행하시오. 이 때 t 통계량의 절대값이 얼마인지 기술하시오.(10점)
- t 통계량은 소수점 다섯째 자리에서 반올림하고 소수점 넷째 자리까지 표시하시오.

```python
from scipy.stats import ttest_rel
stat, p = ttest_rel(usage_gu_total["distance"], usage_gu_total["dist"]) # 2016 vs 2017
round(abs(stat), 4)
```

#### [문제 7] 2018년의 대여소 별 월간 자전거 대여 횟수를 예측해 보기 위해서 다양한 예측 방법을 검토하고 있는데, 본격적인 비교 검토에 앞서, 가장 기본적인 시계열 예측 방법을 테스트해 보기로 하였다. 이를 위해서 가장 대여 횟수가 높았던 대여소 ID 207번 (여의나루역 1번출구 앞)의 2016년~2017년 월별 총 대여 횟수에 대하여 지수 가중 이동평균 평활법(Exponential Weighted Moving Average : EWMA)을 사용하여 월별 총 대여횟수 예측을 수행하고, 가장 마지막 구간인 2017년 12월에 대하여 실제 대여 횟수와 예측된 대여 횟수를 비교하여 평균 절대 백분율 오차(Mean Absolute Percentage Error: MAPE, 단위 %)를 구하시오. 
-	지수 가중 이동평균 평활법(Exponential Weighted Moving Average : EWMA)을 사용 시, Period 값은 3, Alpha 값은 0.9를 사용하시오.
-	2016년/2017년 데이터를 사용하여, 대여소 ID 207번에 대하여 년도월별 대여 횟수를 집계하시오. 
-	2016년 1월부터 2017년 12월까지 순서대로 정렬이 되도록 한다. (특히 Brightics 사용 시)
-	Python인 경우 pandas의 ewm( ).mean( ) method를 사용하시오.
-	평균 절대 백분율 오차(Mean Absolute Percentage Error: MAPE) 값은 소수점 다섯째 자리에서 반올림하고 소수점 넷째 자리까지 표시하시오.(단위(%)는 제외 하고 기술하시오).

(답안예시) 12.3456

```python
usage_2016.columns
```

```python
id207_2016 = usage_2016.loc[usage_2016["id_out"] == 207, :].reset_index(drop = True)
id207_2016.head(2)
```

```python
id207_2016["date_out"] = pd.to_datetime(id207_2016["date_out"])
id207_2016.head(2)
```

```python
type(id207_2016["date_out"][0])
```

```python
id207_2016["year" ] = id207_2016["date_out"].dt.year
id207_2016["month"] = id207_2016["date_out"].dt.month
id207_2016.head()
```

```python
id207_2016_agg = id207_2016[["year", "month"]].groupby(["year", "month"]).agg(len)
id207_2016_agg = id207_2016_agg.to_frame().reset_index()
id207_2016_agg.rename(columns = {0: "count"}, inplace = True)
id207_2016_agg.head()
```

```python
id207_2017 = usage_2017.loc[usage_2017["id"] == 207, :].reset_index(drop = True)
id207_2017.head(2)
```

```python
id207_2017.columns
```

```python
# id207_2017_agg = id207_2017[["date", "count"]].groupby("date").agg("sum")
# id207_2017_agg = id207_2017_agg.reset_index()
id207_2017_agg.columns = ["date", "count"]
id207_2017_agg.head()
```

```python
id207_2017_agg["year"] = id207_2017_agg["date"].astype(str).apply(lambda x: x[:4])
id207_2017_agg["month"] = id207_2017_agg["date"].astype(str).apply(lambda x: x[-2:])

id207 = pd.concat([id207_2016_agg, id207_2017_agg[["year", "month", "count"]]])
id207 = id207.sort_values(["year", "month"]).reset_index(drop = True)
id207
```

```python
#id207["EWMA"] = id207["count"].ewm(min_periods = 3, alpha = 0.9).mean()
id207["EWMA"] = id207["count"].ewm(alpha = 0.9).mean()
y_true = id207.loc[(id207["year"] == "2017") & (id207["month"] == "12"), "count"]
y_pred = id207.loc[(id207["year"] == "2017") & (id207["month"] == "12"), "EWMA"]
#id207
```

```python
round(abs((y_true - y_pred) / y_true) * 100, 4) # MAPE
```

#### [문제 8] 자전거를 대여하고 이용하는 특징이 지역별로 차이가 있는가를 분석해 보고 그 결과에 따라 추가적인 대여소의 설치 또는 자전거 추가 배치 등을 고려하고 있다. 이를 위한 첫 번째 시도로 계층적 군집분석(Hierarchical Clustering)을 하려고 한다. 구체적인 분석 기준은 다음과 같다.
-	2017년 데이터만을 사용하여, 연간 count(횟수), dist(이용거리), duration (사용시간)의 합계를 구별로 집계하시오.
-	군집 방법은 두 클러스터에서 가장 가까운 거리를 사용하는 방식(Single Linkage)을 사용하시오.
-	5개 군집으로 나누어 군집별로 어떤 구가 유사한가를 파악하시오.
-	Normalization을 적용하지 않고 군집분석 시도하시오.
-	R의 경우 TTR 패키지를 사용하시오. 
-	Python인 경우 sklearn.cluster 모듈의 AgglomerativeClustering 클래스를 사용하시오.

 
군집 분석 수행 결과, 종로구와 같은 군집으로 분류된 구를 모두 적으시오.(‘가다나’ 순으로 쉼표로 구분하여 입력)

(답안예시) 강남구,마포구

```python
du_2017 = usage_2017[["count", "dist", "duration", "gu"]].groupby("gu").agg("sum")
du_2017.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 5, linkage = "single").fit(du_2017)
model.labels_
```

```python
du_2017["cluster"] = model.labels_
du_2017
```

```python
du_2017.sort_values("cluster")
```

```python
du_2017.loc[du_2017["cluster"] == 0, :].index
```

#### [문제 9] 다양한 시각으로 분석을 수행하면서 구별로 동질성을 보인다 하더라도 대여소별로 차이가 더 크다는 것을 발견하였다.  대여소가 어떤 지역에 있는지에 따라 시간대별로 자전거 사용 거리의 패턴이 유사한 곳을 군집으로 묶어 분석을 수행하고자 한다.  이 분석에는 k평균 군집법(K Means Clustering)을 사용하기로 하며 다음과 같은 기준을 적용하고자 한다.
-	시간대별 데이터가 있는 2016년 데이터 Set만 사용하시오.
-	대여 시간인 date_out 변수를 이용하여, 24시간으로 표기된 시간 형태인 대여 발생 시간대(0/1/2…/23)를 의미하는 hour 변수를 추가하시오.
-	각 자전거 대여소의 0시부터 23시까지 24개의 시간대별로 대여된 자전거의 사용 거리를 합산하시오.
-	합산 결과 Null이 발생하면 0으로 Replace 하시오.
-	Normalization을 적용하지 않고 군집분석 시도하시오.
-	3개의 클러스터로 군집화하시오. 
-	Seed(Initial Centroid Point)를 값으로 1234 사용하시오.
-	시간을 나타내는 0~23까지의 컬럼명이 순서대로 정렬이 되도록 하시오.
-	R의 경우  Reshape 패키지 사용하시오.

군집 분석된 결과에서, 대여소 ID 207번 (여의나루역 1번출구 앞)이 속한 군집을 찾아 동일 군집에 속한 대여소들의 시간대 1(새벽 1시 이상~새벽 2시 미만)의 자전거의 사용 거리 평균값을 구하시오.
-	사용 거리 평균값은 정수 두 번째 자리 이하는 버리고 세 번째 자리 이상만 기술하시오.

(답안예시) 1234500

```python
usage_2016.head(2)
```

```python
usage_2016.columns
```

```python
usage_2016["hour"] = pd.to_datetime(usage_2016["date_out"], format = "%Y-%m-%d %H:%M").dt.hour

tab = pd.crosstab(index = usage_2016["id_out"], columns = usage_2016["hour"], 
                  values = usage_2016["distance"], aggfunc = "sum").fillna(0)
tab.iloc[:2, :]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 1234).fit(tab)
```

```python
df_results = pd.DataFrame({"id_out": tab.index,
                           "cluster": kmeans.labels_})
df_results.head()
```
